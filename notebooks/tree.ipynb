```python
import matplotlib.pyplot as plt

from raps.plot import FigMeansPlotter

plt.rcParams.update({'font.size': 12})
```

```python
linestyles = ["solid", "dashed", "dotted"]
fig = plt.figure(figsize=(5, 4))
ax = plt.gca()
plt.xlabel(r"$T$")
plt.ylabel("Cumulative Regret")

for i, num_parents in enumerate(range(1, 4)):
  plotter = FigMeansPlotter.from_dir(f"../logdir/tree-p{num_parents}d2n20.01", 
                                     f"{{key}}-p{num_parents}", ax=ax)
  plotter.means(clear_ax=False, linestyle=linestyles[i], lw=3)
```

```python
plt.sca(ax)
plt.savefig("graphics/tree-regret.pdf", format="pdf",
            dpi=1200, bbox_inches="tight")
```
